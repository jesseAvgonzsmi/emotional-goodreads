Merging slightly cleaned and lemmatized reviews

```python
with open (DIR + 'ya_reviews.pkl', 'rb') as f:
    ya_reviews = pickle.load(f)

ya_reviews.head()
```

```python
ya_reviews.info()
```

```python
ya_reviews_lem = pd.read_parquet(DIR +'new_ya_reviews_cleaned_21824.parquet')
ya_reviews_lem.head()
```

```python
ya_reviews_lem.info()
```

```python
#merge on book_id and review_id
ya_reviews_merge = ya_reviews.merge(ya_reviews_lem, on=['book_id', 'review_id'])
```

```python
ya_reviews_merge.info()
```

Finalizing review file

```python
ya_reviews_merge.head()
```

```python
#drop review_text_clean
ya_reviews_merge.drop(columns=['review_text_clean'], inplace=True)
```

```python
#which contain "see full review"
ya_reviews_merge['review_text'].str.contains('see full review').sum()
```

```python
ya_reviews_merge[ya_reviews_merge['review_text'].str.contains('see full review')]
```

```python
import re

ya_reviews_merge['review_text_lemmatized'] = ya_reviews_merge['review_text_lemmatized'].progress_apply(lambda x: re.sub(r'see full review .* https?//\S+', '', x))

ya_reviews_merge['review_text_lemmatized'] = ya_reviews_merge['review_text_lemmatized'].progress_apply(lambda x: re.sub(r'see full review @ katies corner', '', x))

ya_reviews_merge['review_text_lemmatized'] = ya_reviews_merge['review_text_lemmatized'].progress_apply(lambda x: re.sub(r'http\S+', '', x))

ya_reviews_merge['review_text_lemmatized'] = ya_reviews_merge['review_text_lemmatized'].progress_apply(lambda x: re.sub(r'see full review', '', x))
```

```python
ya_reviews_merge[ya_reviews_merge['review_text'].str.contains('see full review')]
```

```python
ya_reviews_merge.to_parquet(DIR + 'ya_reviews_prepped.parquet')
```
