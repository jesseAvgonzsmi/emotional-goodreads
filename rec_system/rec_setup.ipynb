## User Data

```python
# Group by 'user_id' and aggregate 'book_id' into lists
user_read = int_read.groupby('user_id')['book_id'].apply(list).reset_index()
user_read = user_read.rename(columns={'book_id':'read'})
user_reviewed = reviews.groupby('user_id')['book_id'].apply(list).reset_index()
user_reviewed = user_reviewed.rename(columns={'book_id':'reviewed'})
user_interested = int_interested.groupby('user_id')['book_id'].apply(list).reset_index()
user_interested = user_interested.rename(columns={'book_id':'interested'})
user_data = pd.merge(user_read, user_reviewed, on='user_id', how='outer')
user_data = pd.merge(user_data, user_interested, on='user_id', how='outer')
```

```python
user_data['read_len'] = user_data['read'].apply(lambda x: len(x) if isinstance(x, list) else np.nan)
user_data['reviewed_len'] = user_data['reviewed'].apply(lambda x: len(x) if isinstance(x, list) else np.nan)
```

```python
def get_emotions(user_id, rating):
    if isinstance(user_id, list):
        emotions = reviews[(reviews['user_id'].isin(user_id)) & (reviews['rating'] >= rating)][['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust']]
        emotions = emotions.mean()
        return emotions
    else:
        return np.nan
```

```python
user_review_data=reviews.groupby(['user_id','rating']).agg({ 'compassionate':'sum', 'nostalgic':'sum', 'sad':'sum', 'dissapointed':'sum', 'confused':'sum', 'inspired':'sum', 'joyful':'sum','excited':'sum', 'anxious':'sum','sensual':'sum','offended':'sum','optimistic':'sum', 'uncomfortable':'sum'}).reset_index()

user_review_data_2 = reviews.groupby(['user_id','rating']).agg({
    'compassionate': lambda x: (x != 0).sum(),
    'nostalgic': lambda x: (x != 0).sum(),
    'sad': lambda x: (x != 0).sum(),
    'dissapointed': lambda x: (x != 0).sum(),
    'confused': lambda x: (x != 0).sum(),
    'inspired': lambda x: (x != 0).sum(),
    'joyful': lambda x: (x != 0).sum(),
    'excited': lambda x: (x != 0).sum(),
    'anxious': lambda x: (x != 0).sum(),
    'sensual': lambda x: (x != 0).sum(),
    'offended': lambda x: (x != 0).sum(),
    'optimistic': lambda x: (x != 0).sum(),
    'uncomfortable': lambda x: (x != 0).sum(),
    'book_id': 'count'
}).reset_index()

user_review_data_2.columns = ['count_'+str(col) if col != 'user_id' and col != 'rating' else col for col in user_review_data_2.columns]

user_review_data = pd.merge(user_review_data, user_review_data_2, on=['user_id','rating'], how='outer')
```

```python
user_enjoyed = user_review_data[user_review_data['rating'] >= 4].drop('rating', axis=1).groupby('user_id').sum()
user_enjoyed = user_enjoyed[['count_compassionate', 'count_nostalgic',
       'count_sad', 'count_dissapointed', 'count_confused', 'count_inspired',
       'count_joyful', 'count_excited', 'count_anxious', 'count_sensual',
       'count_offended', 'count_optimistic', 'count_uncomfortable',
       'count_book_id']]
for col in user_enjoyed.columns:
    if col != 'count_book_id':
        new_col_name = col.replace('count_', '') + '_p'
        user_enjoyed[new_col_name] = user_enjoyed[col] / user_enjoyed['count_book_id']
user_enjoyed = user_enjoyed[['count_book_id', 'compassionate_p', 'nostalgic_p', 'sad_p',
       'dissapointed_p', 'confused_p', 'inspired_p', 'joyful_p', 'excited_p',
       'anxious_p', 'sensual_p', 'offended_p', 'optimistic_p',
       'uncomfortable_p']]
user_enjoyed.reset_index(inplace=True)
```

```python
user_enjoyed
```

```python
user_disliked = user_review_data[user_review_data['rating'] <= 2].drop('rating', axis=1).groupby('user_id').sum()
user_disliked = user_disliked[['count_compassionate', 'count_nostalgic',
       'count_sad', 'count_dissapointed', 'count_confused', 'count_inspired',
       'count_joyful', 'count_excited', 'count_anxious', 'count_sensual',
       'count_offended', 'count_optimistic', 'count_uncomfortable',
       'count_book_id']]
for col in user_disliked.columns:
    if col != 'count_book_id':
        new_col_name = col.replace('count_', '') + '_p'
        user_disliked[new_col_name] = user_disliked[col] / user_disliked['count_book_id']
user_disliked = user_disliked[['count_book_id', 'compassionate_p', 'nostalgic_p', 'sad_p',
       'dissapointed_p', 'confused_p', 'inspired_p', 'joyful_p', 'excited_p',
       'anxious_p', 'sensual_p', 'offended_p', 'optimistic_p',
       'uncomfortable_p']]
user_disliked.reset_index(inplace=True)
```

```python
for col in user_enjoyed.columns:
    if col != 'user_id':
        user_enjoyed['liked_'+ col] = user_enjoyed[col]
        user_enjoyed.drop(col, axis=1, inplace=True)
```

```python
for col in user_disliked.columns:
    if col != 'user_id':
        user_disliked['disliked_'+ col] = user_disliked[col]
        user_disliked.drop(col, axis=1, inplace=True)
```

```python
#merge user_enjoyed with user_disliked on user_id
user_emotions = pd.merge(user_enjoyed, user_disliked, on='user_id', how='outer')
user_emotions = user_emotions.fillna(0)
```

```python
user_emotions.to_csv(DIR+'user_preferences.csv', index=False)
```
