```python
#open stopwords txt in format word, count as a dataframe
stopwords = pd.read_csv(DIR_new + 'custom_stopwords.txt', sep=', ', header=None, engine='python')
stopwords.columns = ['word', 'count']
```

```python
want_test = yar_word_token[yar_word_token['review_id']=='87e3029d1bdffd25ee370c91d6c6919b']
want_test['tokenized_stop'] = want_test['tokenized'].apply(lambda x: [item for item in x if item not in stopwords['word'].tolist()])
```

```python
#remove any words in word column of stopwords from tokenized column of yar_word_token
yar_word_token['tokenized'] = yar_word_token['tokenized'].progress_apply(lambda x: [item for item in x if item not in stopwords['word'].tolist()])
```

### N-gram

#### bigram (on tokenized words)

```python
from nltk.util import bigrams

yar_bigram = yar_word_token.copy()

yar_bigram['bigrams'] = yar_bigram['tokenized'].progress_apply(lambda x: list(bigrams(x)))
```

```python
yar_bigram
```

#### trigram

```python
from nltk.util import trigrams

yar_trigram = yar_bigram.copy()

yar_trigram['trigrams'] = yar_trigram['tokenized'].progress_apply(lambda x: list(nltk.trigrams(x)))
```
