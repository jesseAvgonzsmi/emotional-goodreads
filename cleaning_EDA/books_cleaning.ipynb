# Cleaning Books file

```python
import gzip
import numpy as np
import pandas as pd
```

## Import, prelim column selection, downsizing types

```python
# Function to process a chunk of data
def process_data(chunk, columns=None):
    # If columns is not None, keep only those columns
    if columns is not None:
        chunk = chunk[columns]
    return chunk

# Function to read data in chunks and process each chunk
def load_data(file_name, head = None, columns=None, chunksize = 1000):
    chunks = []
    count = 0
    with gzip.open(file_name) as fin:
        for chunk in pd.read_json(fin, lines=True, chunksize=chunksize):
            # Process the chunk
            processed_chunk = process_data(chunk, columns)
            chunks.append(processed_chunk)
            
            count += 1
            # break if reaches the head-th chunk
            if (head is not None) and (count > head):
                break

    # Combine all chunks into a single DataFrame
    df = pd.concat(chunks, ignore_index=True)
    
    return df
```

## Formatting / Feature adjustment (not engineering)

Turning complex author lists into plain list of ids

```python
def extract_author_ids(author_list):
    try:
        # Extract the 'author_id' from each dictionary and return a list of ids
        return [d['author_id'] for d in author_list if 'author_id' in d]
    except TypeError:
        return None

ya_books['authors'] = ya_books['authors'].apply(extract_author_ids)
```

Converting popular_shelves to a list of genres (removing the size of shelves) -- they are in order right now from shelves most frequently to least frequently seen on. Eventually woudl like to cross-reference with a list of genres to get a more streamlined list of genres and remove things like 'to-read','currently reading',and 'kindle' which are classifiers but not genres.

```python
def extract_shelves(shelf_list):
    try:
        # Extract the 'shelf name' from each dictionary and return a list of ids
        return [d['name'] for d in shelf_list if 'name' in d]
    except TypeError:
        return None
    
ya_books['popular_shelves'] = ya_books['popular_shelves'].apply(extract_shelves)
```

Making empty series lists NaN

```python
ya_books['series'] = ya_books['series'].apply(lambda x: np.nan if len(x)==0 else x)
```
